```python
##splitting the independent and dependent variable
parnkinsons_data_ind = parkinsons_data.drop("status" , axis=1)
parnkinsons_data_ind

parnkinsons_data_dep = parkinsons_data["status"]
parnkinsons_data_dep
```

```python
## chekcing the target variable
parnkinsons_data_dep.value_counts(normalize=True)
```

```python
plt.figure(figsize=(12,8))
sns.countplot(data = parkinsons_data , x = "status" )
plt.show()
```

```python
### splitting the data into train test split
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(parnkinsons_data_ind ,parnkinsons_data_dep,test_size=0.2,random_state=47,stratify = parnkinsons_data_dep )
```

```python
### MODEL BUILDING
## LOGISTIC REGRESSION
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score , classification_report , confusion_matrix
logit_model = LogisticRegression().fit(x_train , y_train)
logit_model
```

```python
##PREDICT
logit_model_pred_train = logit_model.predict(x_train)
logit_model_pred_test = logit_model.predict(x_test)
```

```python
## EVALUATION
accuracy_train_logit = accuracy_score(y_train , logit_model_pred_train)
accuracy_train_logit
```

```python
accuracy_test_logit = accuracy_score(y_test , logit_model_pred_test)
accuracy_test_logit
```

```python
classification_report_train_logit = classification_report(y_train , logit_model_pred_train)
print(classification_report_train_logit)
```

```python
classification_report_test_logit = classification_report(y_test , logit_model_pred_test)
print(classification_report_test_logit)
```

```python
#using rfe 
from sklearn.feature_selection import RFE

logit_modle_rfe = RFE(logit_model,n_features_to_select=10).fit(x_train , y_train)
logit_modle_rfe
```
